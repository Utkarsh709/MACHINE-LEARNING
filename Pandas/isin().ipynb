```python
import numpy as np
import pandas as pd
```

```python
delivery=pd.read_csv('deliveries.csv')
```

```python
delivery.head(3)
```

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['ball'].count()
x=all_batsman>200
all_batsman[x]
```

```python
all_batsman[x].shape[0]
```

```python
all_batsman[x].index
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['ball'].count()
```

```python
sr=(runs/balls)*100
```
